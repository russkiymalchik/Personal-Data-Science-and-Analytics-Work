
In this dataset, columns as the following:
* Age and sex are self-explanatory
* BMI is body mass index
* BP is average blood pressure
* S1 through S6 are different blood measurements
* Y is the qualitative measure of disease progression over one year

Let's study this dataset using methods of probability and statistics.

### Task 3: What is the the distribution of Age, Sex, BMI and Y variables?

```python
sns.set_style('whitegrid')
def dist(col, bins):
    plt.figure(figsize=(3,5))
    sns.histplot(df[col], kde = True, color ='blue', binwidth=bins)
#     sns.displot(df[col], color ='red', kind="kde")
    plt.show()
    

dist('AGE',2)
dist('SEX',1)
dist('BMI',2)
dist('Y',10)
```

For AGE, SEX, BMI, and Y, the result is quite distinct each other. SEX is a small categorical value, therefore we cannot decide what type of distribution it is. BMI and Y show an indication that they are left skew type of distribution while AGE can be consider normal distribution.
